```python
def nice_max_projection(lund_dataset,index,
                        y_cutoff_bottom=1024,y_cutoff_top=0,
                        x_cutoff_right=512,x_cutoff_left=0):
    import numpy as np
    import matplotlib.pyplot as plt
    t_index = lund_dataset.get_intensity_image(index)

    max_proj = np.max(t_index,axis=0)

    plt.figure(figsize = (10,20))

    plt.imshow(max_proj,cmap = 'gray')
    plt.axis('off')
    
nice_max_projection(wolgast,3)
```

```python
def save_max_proj_in_folder(lund_dataset,index, folder,
                        y_cutoff_bottom=1024,y_cutoff_top=0,
                        x_cutoff_right=512,x_cutoff_left=0):
    import numpy as np
    import matplotlib as plt
    from skimage import io
    
    t_index = lund_dataset.get_intensity_image(index)

    max_proj = np.max(t_index,axis=0)[y_cutoff_top:y_cutoff_bottom,x_cutoff_left:x_cutoff_right]
    
    io.imsave(folder+'max_projection_idx_{}.png'.format(index),max_proj)
    
save_folder = tc.get_folder_path()

save_max_proj_in_folder(wolgast,0,save_folder)
```

```python
for i in range(33):
    save_max_proj_in_folder(wolgast,i,save_folder)
```
